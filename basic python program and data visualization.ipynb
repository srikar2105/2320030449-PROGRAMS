```python
from collections import deque
def bfs(g,s):
  q=deque([s])
  v=set()
  v.add(s)
  while q:
    u=q.popleft()
    print(u,end=" ")
    for i in g[u]:
      if i not in v:
        v.add(i)
        q.append(i)

g={'A':['B','C'],
           'B':['D','E'],
           'C':['F'],
           'D':[],
           'E':[],
           'F':[]
           }
bfs(g,'A')
```

```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount

  def withdraw(self, amount):
    if amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Insufficient funds.")

  def get_balance(self):
    return self.__balance

account = BankAccount(1000)

account.deposit(500)
account.withdraw(200)

print("Current balance:", account.get_balance())
```

```python
from collections import deque

class Graph:
  def __init__(self, graph):
    self.__graph = graph

  def bfs(self, start_node):
    queue = deque([start_node])
    visited = set()
    visited.add(start_node)

    while queue:
      node = queue.popleft()
      print(node, end=" ")

      for neighbor in self.__graph[node]:
        if neighbor not in visited:
          visited.add(neighbor)
          queue.append(neighbor)

graph_data = {
  'A': ['B', 'C'],
  'B': ['D', 'E'],
  'C': ['F'],
  'D': [],
  'E': [],
  'F': []
}
graph = Graph(graph_data)
graph.bfs('A')
```

```python
class Graph:
    def __init__(self, V):
        self.V = V
        self.adj = [[] for _ in range(V)]

    def addEdge(self, v, w):
        self.adj[v].append(w)

    def DLS(self, src, target, limit):
        if src == target:
            return True
        if limit <= 0:
            return False
        for i in self.adj[src]:
            if self.DLS(i, target, limit - 1):
                return True
        return False

    def IDDFS(self, src, target, max_depth):
        for limit in range(max_depth + 1):
            if self.DLS(src, target, limit):
                return True
        return False

V = 7
graph = Graph(V)
graph.addEdge(0, 1)
graph.addEdge(0, 2)
graph.addEdge(1, 3)
graph.addEdge(1, 4)
graph.addEdge(2, 5)
graph.addEdge(2, 6)

src = 0
target = 6
max_depth = 3
if graph.IDDFS(src, target, max_depth):
    print(f"Target {target} is reachable from source {src} within {max_depth} steps.")
else:
    print(f"Target {target} is not reachable from source {src} within {max_depth} steps.")
```

```python
#sub plots

import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2))
plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2))
plt.show()
```
